```python
import pandas as pd

# Load the dataset
df = pd.read_csv('data/retail_sales_dataset.csv')

# Check for missing values
print("Missing values in each column:")
print(df.isnull().sum())

print(df.head())
# Option 2: Fill missing values (example: fill with the mean for numerical columns)
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean())
df['Price per Unit'] = df['Price per Unit'].fillna(df['Price per Unit'].mean())

# Fill categorical missing values with mode
df['Product Category'] = df['Product Category'].fillna(df['Product Category'].mode()[0])

print("Missing values after handling:")
print(df.isnull().sum())
```

```python
print(df.columns)
```

```python
# Convert 'date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')

# Verify the conversion
print(df['Date'].head())
```

```python
# Aggregate data by date to get total quantity sold per day
daily_sales = df.groupby('Date').agg({
    'Quantity': 'sum',
    'Total Amount': 'sum'
}).reset_index()

# Display the aggregated data
print(daily_sales.head())
```

```python
# Aggregate data by product category to get total quantity sold per category
category_sales = df.groupby('Product Category').agg({
    'Quantity': 'sum',
    'Total Amount': 'sum'
}).reset_index()

# Display the aggregated data
print(category_sales.head())
```

```python
# Aggregate data by both date and product category
date_category_sales = df.groupby(['Date', 'Product Category']).agg({
    'Quantity': 'sum',
    'Total Amount': 'sum'
}).reset_index()

# Display the aggregated data
print(date_category_sales.head(10))
```

```python
import matplotlib.pyplot as plt

# Extract month and year from the date
df['Month'] = df['Date'].dt.to_period('M')

# Aggregate sales by month
monthly_sales = df.groupby('Month').agg({
    'Quantity': 'sum',
    'Total Amount': 'sum'
}).reset_index()

# Plot monthly sales trends
plt.figure(figsize=(15, 6))
plt.plot(monthly_sales['Month'].astype(str), monthly_sales['Total Amount'], marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Extract week and year from the date
df['Week'] = df['Date'].dt.to_period('W')

# Aggregate sales by week
weekly_sales = df.groupby('Week').agg({
    'Quantity': 'sum',
    'Total Amount': 'sum'
}).reset_index()

# Plot weekly sales trends
plt.figure(figsize=(15, 6))
plt.plot(weekly_sales['Week'].astype(str), weekly_sales['Total Amount'], marker='o')
plt.title('Weekly Sales Trend')
plt.xlabel('Week')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Scatter plot of price per unit vs quantity sold
plt.figure(figsize=(15, 6))
plt.scatter(df['Price per Unit'], df['Quantity'], alpha=0.5)
plt.title('Relationship Between Price and Quantity Sold')
plt.xlabel('Price per Unit')
plt.ylabel('Quantity Sold')
plt.grid(True)
plt.show()
```

```python
# Calculate the correlation between price per unit and quantity sold
correlation = df['Price per Unit'].corr(df['Quantity'])
print(f'Correlation between Price per Unit and Quantity Sold: {correlation:.2f}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (ensure this matches the structure of your actual dataset)
df = pd.DataFrame({
    'Age': [22, 34, 45, 50, 63, 29, 42, 53, 60, 72],
    'Quantity': [5, 2, 4, 7, 3, 8, 6, 1, 5, 9],
    'Total Amount': [100, 200, 150, 300, 250, 400, 350, 150, 200, 300]
})

# Create age groups
bins = [0, 18, 25, 35, 45, 55, 65, 100]
labels = ['<18', '18-25', '26-35', '36-45', '46-55', '56-65', '65+']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Aggregate data by age group
age_group_sales = df.groupby('Age Group', observed=False).agg({
    'Quantity': 'sum',
    'Total Amount': 'sum'
}).reset_index()

# Plot sales by age group
plt.figure(figsize=(15, 6))
plt.bar(age_group_sales['Age Group'], age_group_sales['Total Amount'])
plt.title('Sales by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Sales Amount')
plt.grid(True)
plt.show()
```

```python
# Aggregate data by gender
gender_sales = df.groupby('Gender').agg({
    'Quantity': 'sum',
    'Total Amount': 'sum'
}).reset_index()

# Plot sales by gender
plt.figure(figsize=(15, 6))
plt.bar(gender_sales['Gender'], gender_sales['Total Amount'], color=['blue', 'pink'])
plt.title('Sales by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Sales Amount')
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Sample data (ensure this matches the structure of your actual dataset)
df = pd.DataFrame({
    'Date': pd.date_range(start='2024-01-01', periods=100, freq='D'),
    'Quantity': range(100),
    'Price per Unit': [10 + x*0.1 for x in range(100)]
})

# Set date as index
df.set_index('Date', inplace=True)

# Calculate moving averages
df['SMA_7'] = df['Quantity'].rolling(window=7).mean()  # 7-day Simple Moving Average
df['SMA_30'] = df['Quantity'].rolling(window=30).mean()  # 30-day Simple Moving Average

print(df.head(10))
```

```python
# Create lag features
df['lag_1'] = df['Quantity'].shift(1)  # 1-day lag
df['lag_7'] = df['Quantity'].shift(7)  # 7-day lag

print(df.head(10))
```

```python
# Calculate price changes
df['price_change'] = df['Price per Unit'].pct_change()  # Percentage change in price

print(df.head(10))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Sample data
df = pd.DataFrame({
    'Date': pd.date_range(start='2024-01-01', periods=100, freq='D'),
    'Quantity': range(100)
})
df.set_index('Date', inplace=True)

# Fit ARIMA model
model = ARIMA(df['Quantity'], order=(5, 1, 0))  # Order is (p, d, q)
model_fit = model.fit()

# Forecast
forecast = model_fit.forecast(steps=10)

# Plot forecast
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Quantity'], label='Historical Data')
plt.plot(pd.date_range(start=df.index[-1], periods=11, freq='D')[1:], forecast, label='Forecast', color='red')
plt.title('ARIMA Forecast')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Fit SARIMA model
model = SARIMAX(df['Quantity'], order=(5, 1, 0), seasonal_order=(1, 1, 1, 7))  # Seasonal order (p, d, q, s)
model_fit = model.fit()

# Forecast
forecast = model_fit.forecast(steps=10)

# Plot forecast
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Quantity'], label='Historical Data')
plt.plot(pd.date_range(start=df.index[-1], periods=11, freq='D')[1:], forecast, label='Forecast', color='red')
plt.title('SARIMA Forecast')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.legend()
plt.show()
```

```python
from prophet import Prophet

# Prepare data for Prophet
df_prophet = df.reset_index().rename(columns={'Date': 'ds', 'Quantity': 'y'})

# Fit Prophet model
model = Prophet()
model.fit(df_prophet)

# Create future dataframe
future = model.make_future_dataframe(periods=10)
forecast = model.predict(future)

# Plot forecast
fig = model.plot(forecast)
plt.title('Prophet Forecast')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Prepare lag features
df['lag_1'] = df['Quantity'].shift(1)
df['lag_7'] = df['Quantity'].shift(7)
df.dropna(inplace=True)

# Prepare training data
X = df[['lag_1', 'lag_7']]
y = df['Quantity']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Fit Random Forest model
model = RandomForestRegressor(n_estimators=100)
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Evaluate
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')

# Plot predictions
plt.figure(figsize=(12, 6))
plt.plot(df.index[-len(y_test):], y_test, label='Actual')
plt.plot(df.index[-len(y_test):], y_pred, label='Predicted', color='red')
plt.title('Random Forest Forecast')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingRegressor

# Fit Gradient Boosting model
model = GradientBoostingRegressor(n_estimators=100)
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Evaluate
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')

# Plot predictions
plt.figure(figsize=(12, 6))
plt.plot(df.index[-len(y_test):], y_test, label='Actual')
plt.plot(df.index[-len(y_test):], y_pred, label='Predicted', color='red')
plt.title('Gradient Boosting Forecast')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
from math import sqrt
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
```

```python
# Sample data
df = pd.DataFrame({
    'Date': pd.date_range(start='2024-01-01', periods=100, freq='D'),
    'Quantity': range(100)
})
df.set_index('Date', inplace=True)

# Ensure frequency is set
df.index.freq = 'D'

# Feature Engineering: Add lag features (optional)
df['lag_1'] = df['Quantity'].shift(1)
df.dropna(inplace=True)

# Split the data
test_size = int(len(df) * 0.2)
train = df[:-test_size]
test = df[-test_size:]

# Train the ARIMA model
model = ARIMA(train['Quantity'], order=(5, 1, 0))
model_fit = model.fit()

# Forecast the test set
predictions = model_fit.forecast(steps=test_size)

# Create a DataFrame for comparison
test['predictions'] = predictions

# Plot actual vs. predicted values
plt.figure(figsize=(12, 6))
plt.plot(train.index, train['Quantity'], label='Train')
plt.plot(test.index, test['Quantity'], label='Test')
plt.plot(test.index, test['predictions'], label='Predicted', color='red')
plt.title('Actual vs Predicted')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.legend()
plt.show()
```

```python
# Assuming you've already trained your model and it's named model_fit

# Number of periods to forecast (e.g., 30 days)
forecast_steps = 30

# Generate forecast
forecast = model_fit.forecast(steps=forecast_steps)

# Create a date range for the forecasted period
forecast_dates = pd.date_range(start=test.index[-1] + pd.Timedelta(days=1), periods=forecast_steps, freq='D')

# Create a DataFrame to hold the forecasted values
forecast_df = pd.DataFrame({'Forecast': forecast}, index=forecast_dates)

# Combine the actual test data with the forecasted values
combined_df = pd.concat([test['Quantity'], forecast_df], axis=0)

print(combined_df)
```

```python
plt.figure(figsize=(14, 7))

# Plot training data
plt.plot(train.index, train['Quantity'], label='Train')

# Plot test data (actuals)
plt.plot(test.index, test['Quantity'], label='Test', color='blue')

# Plot forecast data
plt.plot(forecast_df.index, forecast_df['Forecast'], label='Forecast', color='green')

plt.title('Actual vs Forecast')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Assuming 'test' is your test set containing actual sales data
# Replace the placeholder with the actual data from the 'test' set
actual_future_data = test['Quantity']

# Combine actual and forecasted data into one DataFrame for comparison
comparison_df = pd.DataFrame({
    'Actual': actual_future_data,
    'Forecast': forecast_df['Forecast']
}, index=forecast_dates)

# Plot to compare actual future sales with forecasted sales
plt.figure(figsize=(14, 7))
plt.plot(comparison_df.index, comparison_df['Actual'], label='Actual', color='blue')
plt.plot(comparison_df.index, comparison_df['Forecast'], label='Forecast', color='green')
plt.title('Actual vs Forecasted Sales')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Assuming `forecast_df` and `actual_future_data` are defined earlier
# Check if forecast_df and actual_future_data have matching indices

# Ensure actual_future_data and forecast_df['Forecast'] have the same index
forecast_df = forecast_df.reindex(actual_future_data.index)
comparison_df = pd.DataFrame({
    'Actual': actual_future_data,
    'Forecast': forecast_df['Forecast']
})

# Check if the DataFrame is empty
if comparison_df.empty:
    print("The DataFrame is empty. Check the date range of the forecast and actual data.")
else:
    # Plot to compare actual future sales with forecasted sales
    plt.figure(figsize=(14, 7))
    plt.plot(comparison_df.index, comparison_df['Actual'], label='Actual', color='blue')
    plt.plot(comparison_df.index, comparison_df['Forecast'], label='Forecast', color='green')
    plt.title('Actual vs Forecasted Sales')
    plt.xlabel('Date')
    plt.ylabel('Quantity')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # Calculate MAE and RMSE
    mae = mean_absolute_error(comparison_df['Actual'], comparison_df['Forecast'])
    rmse = np.sqrt(mean_squared_error(comparison_df['Actual'], comparison_df['Forecast']))

    print(f'Mean Absolute Error (MAE): {mae:.2f}')
    print(f'Root Mean Squared Error (RMSE): {rmse:.2f}')
```
